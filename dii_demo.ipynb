```python
import vist
import matplotlib.pyplot as plt
from pprint import pprint
from datetime import datetime
%matplotlib inline
plt.rcParams['figure.figsize'] = (10, 10)
```

```python
vist_images_dir = 'data/val.images'
vist_annotations_dir = 'data'
dii = vist.Description_in_Isolation(vist_images_dir, vist_annotations_dir, ['val'])
```

```python
img_ids = [u'694227468', u'694227412', u'694227488', u'694227508', u'693413687']
dii.show_imgs_with_sents(img_ids)
```

```python
# sents stats
split_to_sent_ids = {'train': [], 'val': [], 'test': []}
for sent in dii.sents:
    album_id = sent['album_id']
    split = dii.Albums[album_id]['split']
    split_to_sent_ids[split] += [sent['id']]
total_sents = 0
for split in ['train', 'val', 'test']:
    print('There are [%s] sents in [%s] split.' % (len(split_to_sent_ids[split]), split))
    total_sents += len(split_to_sent_ids[split])
print('In total, there are [%s] sents.' % (total_sents))
```
