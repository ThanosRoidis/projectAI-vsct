```python
import tensorflow as tf
import time
import pickle
import numpy as np
from scipy.misc import imread, imresize
from data.VGG16.imagenet_classes import class_names
from os import listdir
from os.path import isfile, join
from data.VGG16 import vgg16
```

```python
# init cnn
sess = tf.Session()
imgs = tf.placeholder(tf.float32, [None, 224, 224, 3])
vgg = vgg16.vgg16(imgs, 'data/vgg16/vgg16_weights.npz', sess)
```

```python
start = time.time()

mypath ='D:/gnsid/Documents/MSc/projectAI/val.images/'
images_path = [f for f in listdir(mypath) if isfile(join(mypath, f))]

img2features = dict()

for img_path in images_path:
    print(img_path)
    
    path = mypath + img_path
    # get image features
    img = imread(path, mode='RGB')
    img = imresize(img, (224, 224))
    features = sess.run(vgg.fc2, feed_dict={vgg.imgs: [img]})[0]
    # add image features to dictionary
    img2features[img_path] = features
    
    print('Done')
    
end = time.time();
```

```python
end-start
```

```python
with open('D:/gnsid/Documents/MSc/projectAI/features/img2features', 'wb') as file:
    pickle.dump(img2features, file)
```
